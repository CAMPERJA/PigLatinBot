```python
from transformers import pipeline

pipe = pipeline(
    "text-generation",
    model="microsoft/Phi-3-mini-4k-instruct",
    device_map="auto",
    max_new_tokens=100,
    do_sample=False
)
```

```python
def pig_latin_prompt(user_input):
    return f"""
You are a Pig Latin translator.

Rules:
- If a word starts with a consonant, move the first letter to the end and add "ay"
- If a word starts with a vowel, add "yay" to the end
- Preserve punctuation
- Only return the translated sentence

Translate this:
{user_input}
"""
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def chat():
    print("Pig Latin Bot (type 'quit' to exit)\n")
    print("Please note, responses take quite a bit of time. please be patient.")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "quit":
            break

        prompt = pig_latin_prompt(user_input)

        response = pipe(prompt)[0]["generated_text"]

        answer = response.split("Translate this:")[-1].strip()

        print("Bot:", answer)

chat()
```
